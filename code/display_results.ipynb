```python
import pandas as pd

# Read the results file
df = pd.read_csv('../results/mistral/top_train+prompt.csv')

# Round numeric columns to 3 decimal places
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns
df[numeric_cols] = df[numeric_cols].round(3)

# Remove count and skipped columns
cols_to_drop = [col for col in df.columns if 'count' in col.lower() or 'skipped' in col.lower() or 'pairs' in col.lower()]
df_clean = df.drop(columns=cols_to_drop)

# Display the table
df_clean
```
